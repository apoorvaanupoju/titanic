```python
import pandas as pd
```

```python
train_data=pd.read_csv('titanic_train.csv')
test_data=pd.read_csv('titanic test.csv')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
train_data.isnull().sum()
```

```python
train_data.Age.fillna(train_data.Age.mean(),inplace=True)
test_data.Age.fillna(test_data.Age.mean(),inplace=True)
```

```python
train_data.Cabin.fillna(0,inplace=True)
test_data.Cabin.fillna(0,inplace=True)
```

```python
train_data.Embarked.fillna('S',inplace=True)
```

```python
test_data.Fare.fillna(test_data.Fare.mean(),inplace=True)
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
gender = LabelEncoder()
train_data['Gender'] = gender.fit_transform(train_data['Gender'])

embarked = LabelEncoder()
train_data['Embarked'] = embarked.fit_transform(train_data['Embarked'])
```

```python
sex = LabelEncoder()
test_data['Sex'] = sex.fit_transform(test_data['Sex'])

embarked = LabelEncoder()
test_data['Embarked'] = embarked.fit_transform(test_data['Embarked'])
```

```python
x_train=train_data.drop(['Name','Ticket','Cabin','Survived'],axis=1)
y_train=train_data['Survived']
x_test=test_data.drop(['Name','Ticket','Cabin'],axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
```

```python
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)
y_pred
```

```python
feature_imp=pd.DataFrame()
feature_imp['col']=x_train.columns
feature_imp['imp']=rf.feature_importances_*100
feature_imp.sort_values('imp',ascending=False)
```

```python
x_train=x_train.drop(['Embarked','Parch'],axis=1)
x_test=x_test.drop(['Embarked','Parch'],axis=1)
```

```python
param_grid = { "n_estimators"      : [100,50,150],             
               "max_features"      : [1,2,3,4,5],     
               "min_samples_leaf"  : [3,5,8,10],
               "criterion"     :['gini','entropy'],
              "max_depth":[6,8,7,9,10,12]
                }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
optimized_tree=GridSearchCV(rf,param_grid,scoring='accuracy',verbose=2,cv=5)
```

```python
optimized_tree.fit(x_train,y_train)
```

```python
optimized_tree.best_params_
```

```python
optimized_tree.best_score_
```

```python
final_model=optimized_tree.best_estimator_
```

```python
final_model.fit(x_train,y_train)
```

```python
y_predict=final_model.predict(x_test)
```

```python
Final_submission = pd.DataFrame()
Final_submission['PassengerId'] = test_data['PassengerId']
Final_submission['Survived'] = y_predict
Final_submission
```

```python
Final_submission.to_csv('Final_submission(RF_Titanic).csv', index=False)
```
